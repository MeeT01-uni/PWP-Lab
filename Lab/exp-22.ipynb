```python
import pandas as pd

# Load data from CSV
csv_file_path = 'nba.csv'
df_csv = pd.read_csv(csv_file_path)

print("CSV Data:")
print(df_csv)
```

```python
# Check data types of each column
print(df_csv.dtypes)
```

```python
# Fill missing values in the Salary column with the mean
df_csv['Salary'].fillna(df_csv['Salary'].mean(), inplace=True)

print("Missing values in Salary column after filling:")
print(df_csv['Salary'].isnull().sum())
print("\nFirst few rows of Salary column:")
print(df_csv['Salary'].head(10))
```
